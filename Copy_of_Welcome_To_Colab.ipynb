## Select and Train Forecasting Model

### Subtask:
Choose an appropriate forecasting model (e.g., SARIMA, Prophet) and train it using the prepared sales data. This step will involve splitting the data into training and testing sets to evaluate the model's performance.


**Reasoning**:
The subtask requires splitting the `monthly_sales_agg` DataFrame into training and testing sets, importing the `SARIMAX` model, defining its order, and fitting it to the training data. I will start by splitting the data, reserving the last 12 months for the test set.



```
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Split data into training and testing sets (last 12 months for testing)
split_point = len(monthly_sales_agg) - 12
train_data = monthly_sales_agg.iloc[:split_point]
test_data = monthly_sales_agg.iloc[split_point:]

# Define SARIMA model orders
order = (1, 1, 1)        # Non-seasonal (p,d,q)
seasonal_order = (1, 1, 1, 12) # Seasonal (P,D,Q,S)

# Fit the SARIMAX model
model = SARIMAX(train_data['Sales'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
model_fit = model.fit(disp=False)

print("SARIMAX model trained successfully.")
print(model_fit.summary())
```

**Reasoning**:
The previous execution resulted in warnings regarding frequency information and insufficient observations for seasonal ARMA parameters. Explicitly setting the frequency to 'ME' (Month End) in the `SARIMAX` model constructor will address the frequency warning and improve model specification, although the issue of too few observations for robust seasonal parameter estimation might persist due to data length.



```
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Split data into training and testing sets (last 12 months for testing)
split_point = len(monthly_sales_agg) - 12
train_data = monthly_sales_agg.iloc[:split_point]
test_data = monthly_sales_agg.iloc[split_point:]

# Define SARIMA model orders
order = (1, 1, 1)        # Non-seasonal (p,d,q)
seasonal_order = (1, 1, 1, 12) # Seasonal (P,D,Q,S)

# Fit the SARIMAX model, explicitly setting the frequency
model = SARIMAX(train_data['Sales'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False, freq='ME')
model_fit = model.fit(disp=False)

print("SARIMAX model trained successfully.")
print(model_fit.summary())
```

**Reasoning**:
The previous execution still raised a `ValueWarning` regarding frequency information, even though `freq='ME'` was passed to the `SARIMAX` constructor. This indicates that the index of the DataFrame itself needs to have the frequency set. I will explicitly set the frequency of the `train_data` index to 'ME' before passing it to the `SARIMAX` model to resolve this warning and ensure proper frequency handling.



```
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Split data into training and testing sets (last 12 months for testing)
split_point = len(monthly_sales_agg) - 12
train_data = monthly_sales_agg.iloc[:split_point]
test_data = monthly_sales_agg.iloc[split_point:]

# Ensure the index has a defined frequency
train_data = train_data.asfreq('ME')

# Define SARIMA model orders
order = (1, 1, 1)        # Non-seasonal (p,d,q)
seasonal_order = (1, 1, 1, 12) # Seasonal (P,D,Q,S)

# Fit the SARIMAX model
model = SARIMAX(train_data['Sales'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
model_fit = model.fit(disp=False)

print("SARIMAX model trained successfully.")
print(model_fit.summary())
```

**Reasoning**:
The previous steps successfully trained the SARIMAX model, although with a warning about too few observations for seasonal parameters. The next logical step in time series analysis is to generate predictions (forecasts) using the trained model.



```
import matplotlib.pyplot as plt

# Generate predictions for the test set
start_index = len(train_data)
end_index = len(monthly_sales_agg) - 1
predictions = model_fit.predict(start=start_index, end=end_index)

# Print the predictions and actual values for comparison
print("Predictions:")
print(predictions)
print("\nActual values (test_data):")
print(test_data['Sales'])

# Plot the training data, test data, and predictions
plt.figure(figsize=(12, 6))
plt.plot(train_data['Sales'], label='Training Data')
plt.plot(test_data['Sales'], label='Actual Sales (Test Data)', color='orange')
plt.plot(predictions, label='SARIMAX Predictions', color='green', linestyle='--')
plt.title('SARIMAX Model: Training, Actual, and Predicted Sales')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

## Generate Forecast and Extract Components

### Subtask:
Generate future sales forecasts using the trained model and extract key components such as trend and seasonality through time series decomposition.


**Reasoning**:
First, I will generate future sales forecasts for the next 12 months using the trained SARIMAX model and calculate the confidence intervals.



```
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose

# Generate future sales forecasts for the next 12 months
forecast_steps = 12
forecast_result = model_fit.get_forecast(steps=forecast_steps)
forecast_mean = forecast_result.predicted_mean
forecast_conf_int = forecast_result.conf_int()

print("Future sales forecasts (mean):")
print(forecast_mean)
print("\nForecast confidence intervals:")
print(forecast_conf_int)
```

**Reasoning**:
Now that forecasts are generated, I will perform time series decomposition on the `monthly_sales_agg` DataFrame to extract its trend, seasonal, and residual components. I will use a seasonal period of 12 (for monthly data) and an additive model, which is suitable when the magnitude of the seasonal fluctuations is constant over time.



```
decomposition = seasonal_decompose(monthly_sales_agg['Sales'], model='additive', period=12)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

print("Time series decomposition performed successfully. Trend, seasonal, and residual components are extracted.")
```

**Reasoning**:
I will create a comprehensive plot to visualize the historical sales data, the SARIMAX model's predictions for the test set, future forecasts, their confidence intervals, and the decomposed trend and seasonal components. This will provide a complete overview of the time series analysis.



```
import matplotlib.pyplot as plt
import pandas as pd

# Extend monthly_sales_agg to include the forecast index for plotting consistency
full_sales_data = pd.concat([monthly_sales_agg['Sales'], forecast_mean])

plt.figure(figsize=(16, 12))

# Plot historical sales and forecasts
plt.subplot(4, 1, 1)
plt.plot(monthly_sales_agg['Sales'], label='Historical Sales')
plt.plot(predictions, label='SARIMAX Test Predictions', color='orange', linestyle='--')
plt.plot(forecast_mean, label='SARIMAX Future Forecast', color='green')
plt.fill_between(forecast_conf_int.index, forecast_conf_int.iloc[:, 0], forecast_conf_int.iloc[:, 1], color='green', alpha=0.2, label='Confidence Interval')
plt.title('Retail Sales: Historical, Predicted, and Forecasted with Confidence Intervals')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)

# Plot Trend Component
plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend Component', color='red')
plt.ylabel('Trend')
plt.legend()
plt.grid(True)

# Plot Seasonal Component
plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonal Component', color='blue')
plt.ylabel('Seasonality')
plt.legend()
plt.grid(True)

# Plot Residual Component
plt.subplot(4, 1, 4)
plt.plot(residual, label='Residual Component', color='purple')
plt.ylabel('Residuals')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
